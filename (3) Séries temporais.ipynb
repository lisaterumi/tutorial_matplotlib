# Séries temporais por gráfico de linha (Time Series by line plot)

A série temporal é um gráfico de linha que conecta pontos de dados com uma linha reta. É útil para entender a tendência ao longo do tempo, ajudando a explicar a correlação entre os pontos pela tendência. 

Uma tendência ascendente significa correlação positiva enquanto a tendência descendente significa correlação negativa. É usado principalmente em modelos de previsão e monitoramento.

Quando usar: Séries temporais devem ser usadas quando variáveis únicas ou múltiplas devem ser plotadas ao longo do tempo, como análise do mercado de ações de empresas e previsão do tempo.

### Importando bibliotecas e dados

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

stock_data = pd.read_csv('datasets/stocks.csv')
````

````python
stock_data.head()
````

Primeiro, convertemos ```Date``` em um objeto *pandas DateTime* para facilitar a plotagem de dados.

````python
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.head()
````

#### Comparar os dados lado a lado

````python
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.set_title('AAPL vs IBM(inset)')

ax1.plot(stock_data['Date'],
         stock_data['AAPL'],
         color='green')

ax2.plot(stock_data['Date'],
         stock_data['IBM'], 
         color='blue')
plt.show()
````

```fig.add_axes``` é usado para plotar a tela ([ver aqui as diferenças entre ```add_axes``` e ```add_subplot```](https://stackoverflow.com/questions/43326680/what-are-the-differences-between-add-axes-and-add-subplot)).

```plt.plot()``` recebe o primeiro argumento como dados numéricos (data) e o segundo argumento, dados numéricos do valor da ação. 

#### Comparar vários dados

````python
# Figsize para largura e altura do grafico
fig = plt.figure(figsize=(15,7))

fig.suptitle('Comparação de preços de ações 2007-2017', 
             fontsize=20)

ax1 = fig.add_subplot(231)
ax1.set_title('MSFT')

ax1.plot(stock_data['Date'],
         stock_data['MSFT'], 
         color='green')

ax2 = fig.add_subplot(232)
ax2.set_title('GOOG')

ax2.plot(stock_data['Date'],
         stock_data['GOOG'], 
         color='purple')

ax3 = fig.add_subplot(233)
ax3.set_title('SBUX')

ax3.plot(stock_data['Date'],
         stock_data['SBUX'], 
         color='magenta')

ax4 = fig.add_subplot(234)
ax4.set_title('ADBE')

ax4.plot(stock_data['Date'],
         stock_data['ADBE'], 
         color='orange')

ax4 = fig.add_subplot(235)
ax4.set_title('NFLX')

ax4.plot(stock_data['Date'],
         stock_data['NFLX'], 
         color='chocolate')

ax4 = fig.add_subplot(236)
ax4.set_title('ORCL')

ax4.plot(stock_data['Date'],
         stock_data['ORCL'], 
         color='teal')

plt.show()
````

```add_axes``` adiciona eixos a uma figura, enquanto ```add_subplot``` adiciona vários subplots a uma figura. 

```fig.add_subplot(237)``` não pode ser executado porque existem apenas 6 *canvas* possíveis.

Podemos ver que as ações das empresas de tecnologia estão seguindo uma tendência de alta, mostrando resultados positivos para os comerciantes investirem em ações.
