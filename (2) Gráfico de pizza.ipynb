# Gráfico de pizza (*Pie Chart*)

É um gráfico circular dividido em fatias para ilustrar a proporção numérica. A fatia de um gráfico de pizza mostra a proporção das partes de um todo.

Quando usar: evite usar o gráfico de pizza, pois é difícil comparar as seções do gráfico. O gráfico de barras deve usado em vez disso, pois a comparação de seções é fácil (veja mais [aqui](https://www.data-to-viz.com/caveat/pie.html)).

### Importando bibliotecas e dados

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

t_mov= pd.read_csv('datasets/types_movies.csv')
````

````python
t_mov
````

O gráfico de pizza mostra a distribuição de dados com base na proporção da pizza ocupada.

````python
plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'])

plt.axis('equal')

plt.show()
````

```plt.pie()``` usa os dados numéricos como primeiro argumento, ou seja, porcentagem e rótulos a serem exibidos como segundo argumento (Setor). Em última análise, mostra a distribuição dos dados em proporção à pizza.

### Componentes do gráfico de pizza
* <b>wedges</b> Lista das fatias do gráfico
* <b>texts:</b> Lista de textos que representam os rótulos
* <b>autotexts:</b> Lista dos valores numéricos (disponível apenas se o valor de "autopct" para o gráfico de pizza não for "None")

````python
wedges, texts, autotexts = plt.pie(t_mov['Percentage'], 
                                    labels=t_mov['Sector'],
                                    autopct='%.2f')

plt.axis('equal')

print('Fatias: ', wedges)
print('Textos: ', texts)
print('Autotextos: ', autotexts)
````

Podemos ver os componentes que fazem um gráfico de pizza, que retorna as fatias, os rótulos dos textos e assim por diante.

#### Personalizações de gráfico de pizza
  Definimos os valores das cores e propriedades "autopct" (formato dos valores a serem exibidos).

Um gráfico de pizza pode ser facilmente personalizado e os valores de cores e rótulos são formatados.

````python
colors = ['darkorange', 'sandybrown', 'darksalmon', 'orangered','chocolate']

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f')

plt.axis('equal')

plt.show()
````

#### A propriedade "explodir"
Para destacar uma fatia específica do gráfico de pizza, usamos "explode" para separá-la do resto do gráfico.

O valor para "explode" representa a fração do raio com a qual compensar cada fatia.

````python
explode = (0, 0.1, 0, 0, 0.3)

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
````

```explode``` é usado para separar os pontos da pizza, como uma pizza sendo cortada.
