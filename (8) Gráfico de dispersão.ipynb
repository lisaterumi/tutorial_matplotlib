# Gráfico de dispersão (*Scatter Plot*)

O gráfico de dispersão ajuda a visualizar 2 variáveis numéricas e a identificar a relação dos dados com cada variável, ou seja, padrões de correlação ou tendência. Também ajuda na detecção de *outliers* no gráfico.

Quando usar: é usado em conceitos de aprendizado de máquina como regressão, onde x e y são variáveis contínuas. Ele também é usado em dispersões de *cluster* ou na detecção de *outlier*.


### Importando bibliotecas e dados

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

countries=pd.read_csv('datasets/countries.csv')
````

````python
countries.head()
````

````python
data_2007=countries[countries['year']==2007]
````

## Plotando o gráfico

````python
plt.scatter(data_2007['gdpPerCapita'],data_2007['lifeExpectancy'])
plt.title('Expectativa de vida vs PIB per capita em 2007 ')
plt.xlabel('PIB per capita ($)')
plt.ylabel('Expectativa de vida')
plt.show()
````

````python
data_2007['gdpPerCapita'].corr(data_2007['lifeExpectancy'])
````

```plt.scatter()``` leva 2 argumentos numéricos para espalhar pontos de dados no gráfico. É semelhante ao gráfico de linha, mas sem as linhas retas conectadas. *corr* é a correlação, ou seja, o quanto o PIB está correlacionado com a expectativa de vida. 

Como podemos ver que é positivo, significa que à medida que o PIB de um país aumenta, a expectativa de vida também aumenta.

````python
plt.scatter(np.log10(data_2007['gdpPerCapita']),data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
````

````python
np.log10(data_2007['gdpPerCapita']).corr(data_2007['lifeExpectancy'])
````

Ao analisar o *log* do PIB, podemos ver que há uma correlação muito melhor, pois podemos ajustar os pontos melhor, ele converte o PIB na escala logarítmica, ou seja, log ($ 1000) = 3.

## Gráfico de dispersão 3D

O Gráfico de dispersão 3D (ou *3D Scatterplot*) ajuda a visualizar 3 variáveis numéricas em um gráfico tridimensional.

````python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(countries['year'],countries['gdpPerCapita'],countries['lifeExpectancy'], s=30)
plt.show()
````

Aqui adicionamos 3 variáveis numéricas desta vez, o que nos permite inferir que, à medida que o ano e o PIB aumentam, a expectativa de vida também aumenta.
